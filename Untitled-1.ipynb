```python
import pandas as pd
import glob
```

```python

imgs_path = 'data/images/'
csv_path  = r'data\data_labels.csv'
file_list = glob.glob(imgs_path + "*")
print(file_list)

df = pd.read_csv(csv_path,index_col ="filename")

data = []

for class_path in file_list:
    for img_path in glob.glob(class_path+'/*.jpg'):
        file_name = img_path.split("\\")[-1]

        data.append([img_path, df.loc[file_name,'x']])
print(data)
```
